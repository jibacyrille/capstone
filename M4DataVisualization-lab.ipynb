# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

import matplotlib.pyplot as plt
QUERY = """
SELECT ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Plot a histogram of the 'ConvertedComp' column
plt.figure(figsize=(12, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Plot a box plot of the 'Age' column
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'].dropna(), vert=False, patch_artist=True)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Plot a scatter plot of 'Age' vs 'WorkWeekHrs'
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], color='blue', alpha=0.5)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Create a bubble plot of 'WorkWeekHrs' vs 'CodeRevHrs' with 'Age' as bubble size
plt.figure(figsize=(12, 8))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""
df = pd.read_sql_query(QUERY, conn)

# Count the occurrences of each database
top_databases = df['DatabaseDesireNextYear'].value_counts().head(5)

# Plot a pie chart of the top 5 databases
plt.figure(figsize=(10, 8))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()


QUERY = """
SELECT *
FROM LanguageDesireNextYear
"""
df = pd.read_sql_query(QUERY, conn)

# Count the occurrences of each database
top_databases = df['LanguageDesireNextYear'].value_counts().head(5)

# Plot a pie chart of the top 5 databases
plt.figure(figsize=(10, 8))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Languages Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()


QUERY = """
SELECT *
FROM LanguageWorkedWith
"""
df = pd.read_sql_query(QUERY, conn)

# Count the number of respondents who indicated 'SQL' in the 'LanguageWorkedWith' column
sql_count = df['LanguageWorkedWith'].str.contains('SQL', na=False).sum()
print(sql_count)


QUERY = """
SELECT *
FROM DatabaseWorkedWith
"""
df = pd.read_sql_query(QUERY, conn)

# Count the number of respondents who indicated 'MySQL' in the 'LanguageWorkedWith' column
sql_count = df['DatabaseWorkedWith'].str.contains('MySQL', na=False).sum()
print(sql_count)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Filter the data for respondents aged 30 to 35
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median 'WorkWeekHrs' and 'CodeRevHrs' for this age group
median_workweekhrs = filtered_df['WorkWeekHrs'].median()
median_coderevhrs = filtered_df['CodeRevHrs'].median()

# Create a DataFrame for the median values
median_data = pd.DataFrame({
    'Median Hours': [median_workweekhrs, median_coderevhrs],
    'Category': ['WorkWeekHrs', 'CodeRevHrs']
})

# Create a stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(median_data['Category'], median_data['Median Hours'], color=['skyblue', 'orange'])
plt.xlabel('Category')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Query the 'ConvertedComp' and 'Age' columns from the 'master' table
QUERY = """
SELECT ConvertedComp, Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Filter the data for respondents aged 45 to 60
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate the median 'ConvertedComp' for each age
median_convertedcomp = filtered_df.groupby('Age')['ConvertedComp'].median()

# Create a line chart
plt.figure(figsize=(12, 6))
plt.plot(median_convertedcomp.index, median_convertedcomp.values, marker='o', color='skyblue', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()


QUERY = """
SELECT ConvertedComp, Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Filter the data for respondents aged 45 to 60
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

# Calculate the median 'ConvertedComp' for each age
median_convertedcomp = filtered_df.groupby('Age')['ConvertedComp'].median()

# Create a line chart
plt.figure(figsize=(12, 6))
plt.plot(median_convertedcomp.index, median_convertedcomp.values, marker='o', color='skyblue', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 25 to 30')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Calculate the count of each category in the 'MainBranch' column
mainbranch_counts = df['MainBranch'].value_counts()

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Count of Respondents by MainBranch')
plt.show()
```
